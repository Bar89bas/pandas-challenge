```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
#  Calculating the passing maths score
score_passing_math = school_data_complete[school_data_complete["maths_score"]>=50]
```

```python
# percentage of passing maths for each school 
maths_passing_percent = score_passing_math.groupby(["school_name"]).count()["student_name"]/total_students*100
maths_passing_percent
```

```python
# Calculating passing reading score 
score_passing_reading = school_data_complete[school_data_complete["reading_score"]>=50]
```

```python
# percentage of passing reading for each school 
reading_passing_percent = score_passing_reading.groupby(["school_name"]).count()["student_name"]/total_students*100
reading_passing_percent
```

```python
# Calculating passing overall maths and reading score
score_passing_both = school_data_complete[(school_data_complete["reading_score"]>=50) &
                                         (school_data_complete["maths_score"]>=50)]
```

```python
# percentage of passing overall maths and reading 
overall_passing_percentage = score_passing_both.groupby(["school_name"]).count()["student_name"]/total_students*100
overall_passing_percentage
```

```python
# Convert to DataFrame
per_school_summary = pd.DataFrame({"School Type": school_types,
                                   "Total Students": total_students,
                                   "Total School Budget": school_budget,
                                   "Per Student Budget": school_capital,
                                   "Average Maths Score": school_maths_score,
                                   "Average Reading Score": school_reading_score,
                                   "% Passing Maths": maths_passing_percent,
                                   "% Passing Reading": reading_passing_percent,
                                   "% Overall Passing": overall_passing_percentage})

# Minor data wrangling
per_school_summary = per_school_summary[["School Type", "Total Students", "Total School Budget", "Per Student Budget",
                                         "Average Maths Score", "Average Reading Score", 
                                         "% Passing Maths", "% Passing Reading", 
                                         "% Overall Passing"]]
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

# Display the DataFrame
per_school_summary
```
